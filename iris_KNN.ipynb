```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
iris.keys()
```

```python
print(iris['DESCR'])
```

```python
print(iris['feature_names'])
```

```python
print(iris['target'])
```

```python
data = pd.DataFrame(iris['data'], columns = iris['feature_names'])
data
```

```python
y = pd.DataFrame(iris['target'], columns = ['target_names'])
y
```

```python
data = pd.concat([data, y], axis = 1)
data
```

```python
def species(x):
    if x == 0:
        value = "setosa"
    elif x == 1:
        value = "versicolor"
    else:
        value = "virginica"
    return value
```

```python
data['species'] = data['target_names'].apply(species)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data_pivottable = pd.pivot_table(data, index = ['species'])
print(data_pivottable)
```

```python
x = data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = data[['species']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
# the accurancy of the test
knn.score(x_test, y_test)
```

```python
# predict a new data which species belong for
new = np.array([[5, 2.9, 1, 0.2]])
predict = knn.predict(new)
print("The predict species is:", predict)
```
